```python
def phone_block(x):
    if len(str(x)) >= 10:
        return str(x)[:2]
    else:
        return 0

def full_name_block(x):
    return str(x).split(' ')[0][:3]
```

```python
df1_clean['phone_block'] = df1_clean['phone'].apply(phone_block)
df1_clean['full_name_block'] = df1_clean['full_name'].apply(full_name_block)
df1_clean.head()
```

```python
def get_embedding(text):
    return model.encode(text)

def process_group(group):
    # group['full_name_embedding'] = group['full_name'].apply(get_embedding)
    group['phone_embedding'] = group['phone'].apply(get_embedding)
    group['email_embedding'] = group['email'].apply(get_embedding)
    # group['address_embedding'] = group['address'].apply(get_embedding)
    # group['birthdate_embedding'] = group['birthdate'].apply(get_embedding)

    group['combined_embedding'] = group.apply(
        lambda row: np.concatenate(
            [row['email_embedding'], row['phone_embedding']]
        ), axis=1
    )

    embeddings = np.vstack(group['combined_embedding'].values)
    similarity_matrix = cosine_similarity(embeddings)

    threshold = 0.8

    group['similar_uids'] = [[] for _ in range(len(group))]

    group_indices = group.index.tolist()

    for i in range(len(group)):
        for j in range(i + 1, len(group)):  # Сравниваем только правую часть матрицы
            if similarity_matrix[i, j] > threshold:
                uid_1 = group.iloc[i]['uid']
                uid_2 = group.iloc[j]['uid']
                
                group.at[group_indices[i], 'similar_uids'].append(uid_1)
                group.at[group_indices[i], 'similar_uids'].append(uid_2)
                group.at[group_indices[j], 'similar_uids'].append(uid_1)
                group.at[group_indices[j], 'similar_uids'].append(uid_2)
                
    return group[['full_name', "email", 'phone', 'birthdate', 'uid', 'similar_uids']]
```
